```python
import pandas as pd
import numpy as np
import json
from bs4 import BeautifulSoup
from urllib import request
```

```python
apartments = pd.read_csv("../../Data/kursk_apartments.csv", index_col=0)
apartments.head()
```

```python
def get_index_by_street(street):
    street_converted = request.quote(street)
    url = f'https://indexphone.ru/?parent=%D0%9A%D1%83%D1%80%D1%81%D0%BA%D0%B0%D1%8F+%D0%BE%D0%B1%D0%BB%D0%B0%D1%81%D1%82%D1%8C%2C+%D0%B3.+%D0%9A%D1%83%D1%80%D1%81%D0%BA&search={street_converted}'
    fp = request.urlopen(url)
    mybytes = fp.read()

    mystr = mybytes.decode("utf8")
    fp.close()
    return(mystr)
```

```python
def find_index(parser):
    try:
        return parser.find("span", attrs={"class": "post-search-list-postalcode"}).text
    except AttributeError as ex:
        return None
```

```python
indexes = dict()
for street in apartments['Улица'].unique():
    html_string = get_index_by_street(street)
    parser = BeautifulSoup(html_string, 'html.parser')
    indexes[street] = find_index(parser)
    print(f'"{street}": {indexes[street]}')
```

```python
indexes['Союзная ул.'] = 305044
indexes['Республиканская ул.'] = 305044
indexes['Дружининская ул.'] = 305001
indexes['Майский б-р'] = 305005
indexes['Чумаковская ул.'] = 305035
indexes['ул. Степана Разина'] = 305004
indexes['Сумская ул.'] = 305007
```

```python
with open("locations_postalcodes.json", 'w', encoding='utf8') as file:
    json.dump(indexes, file, ensure_ascii=False)
```
